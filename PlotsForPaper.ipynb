```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lamb = lambda x,y,z: x**2 + y**2 + z**2 -2*x*y - 2*y*z - 2*x*z
f1 = lambda x,M,m1,m2,m3: m1**2 + m3**2 +((M**2 - x - m1**2)*(x + m3**2 - m2**2) + np.sqrt(lamb(x,m2**2,m3**2)*lamb(x,M**2,m1**2)))/(2*x)
f2 = lambda x,M,m1,m2,m3: m1**2 + m3**2 +((M**2 - x - m1**2)*(x + m3**2 - m2**2) - np.sqrt(lamb(x,m2**2,m3**2)*lamb(x,M**2,m1**2)))/(2*x)
```

```python
x = np.linspace(0, 5, 500)
xdots = np.random.uniform(0,3,500)
ydots = np.random.uniform(0,3,500)
realxdots = []
realydots = []

M = 1.864
m1 = 0.13957018
m2 = 0.13957018
m3 = 0.1349766
newx = []
y1 = []
y2 = []

for i in x:
    if lamb(i,M,m1) > 0 or lamb(i,m2,m3) > 0:
        newx.append(i)
        y1.append(f1(i,M,m1,m2,m3))
        y2.append(f2(i,M,m1,m2,m3))
        
for i in range(len(xdots)):
    if ydots[i] < f1(xdots[i],M,m1,m2,m3) and ydots[i] > f2(xdots[i],M,m1,m2,m3):
        realxdots.append(xdots[i])
        realydots.append(ydots[i])

plt.plot(newx, y1, label = "Min")
plt.plot(newx, y2, label= "Max")
plt.scatter(realxdots,realydots)

plt.legend()
plt.xlim(0,3.2)
plt.ylim(0,3.2)

plt.xlabel("$m_{23}^2$", fontsize=20)
plt.ylabel("$m_{13}^2$", fontsize = 20)

plt.show()
```
